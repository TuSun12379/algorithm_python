```python
import matplotlib.pyplot as plt
import numpy as np

def f1(x):
    return 8*x



def f3(x):
    result = 2*x**2
    return result

def f4(x):
    result = x**3
    return result

def f5(x):
    result = 2**x
    return result
    

x = np.arange(0.1, 100, 1)
l1 = plt.plot(x, f1(x), 'bo')
plt.setp(l1, markersize = 1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f2(x):
    result = 4*x*np.log2(x)
    return result

x = np.arange(0.1, 100, 1)
l1 = plt.plot(x, f2(x), 'bo')
plt.setp(l1, markersize = 1)
plt.show()
```
